    - python 에서는 maplotlib

```python
from matplotlib import pyplot as plt
plt.plot([1,2,3,4,5],[1,4,8,16,25])
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np
x = np.arange(1,6)
y = x ** 2
plt.plot(x,y)
plt.show()
```

arange() 함수

    - 그래프의 x 와 y 의 범위를 설정하는데 사용, range() 함수와 유사함
    - 다만 range() 는 리스트를 생성, arange() 는 배열을 생성

numpy 객체.arange(start, stop, step)

    - start : 생성할 배열의 시작 숫자, 생략할 경우 0 부터 시작
    - stop  : 배열 생성을 중지하는 숫자, 제시된 정지값의 이전 값(stop-1)까지 생성
    - step  : 출력할 숫자 사이의 간격, 생략할 경우 1

코드 예시

    - np.arange(5)      -> array([0,1,2,3,4])
    - np.arange(2,5)    -> array([2,3,4])
    - np.arange(1,5,2)  -> array([1,3])

```python
# 산점도

from matplotlib import pyplot as plt 
import numpy as np

x = np.arange(1,6)
y = np.arange(3,8)
plt.scatter(x,y)
plt.show()
```

```python
# 막대그래프

from matplotlib import pyplot as plt 
import numpy as np
x = ['mon', 'tue', 'wed', 'thu', 'fri']
y = [50, 40, 20, 60, 35]
plt.bar(x,y)
plt.show()
```

```python
# 파이차트
# pyplot 객체.pie(영역별 비율, labels=영역별 레이블, autopct='표시할 숫자 형식')
# autopct : 부채꼴 안에 표시할 비율의 형식으로 소수점 아래 첫째 자리까지 표현하려면 '%1.1f%%' 작성

from matplotlib import pyplot as plt 
import numpy as np
ratio = [20, 30, 40, 70]
label = ['A', 'B', 'C', 'D']
plt.pie(ratio, labels = label, autopct = '%1.1f%%')
plt.show()
```

```python
# 히스토그램
# pyplot 객체.hist(데이터리스트, bins = 가로축 구간의 개수)
# bins 기본값 : 10

from matplotlib import pyplot as plt 
import numpy as np
data = [3,4,5,5,5,6,7,8,3,2,7,7,7,7,8,9,10,10,9]
plt.hist(data, bins = 30)
plt.show()
# 2 -> 1개
# 3 -> 2개
# 4 -> 1개
# 5 -> 3개
```

```python
# 옵션
# pyplot 객체.그래프명(..., color='색상', linestyle='선기호')
# pyplot 객체,xlabel('레이블명', fontsize=글자크기, labelad=여백크기)
# pyplot 객체,ylabel('레이블명', fontsize=글자크기, labelad=여백크기)

x = np.arange(1,7)
y = np.arange(2,8)
plt.plot(x,y, color='red', marker='D', linestyle='--')
plt.xlabel('X lable', fontsize=15, labelpad=20)
plt.ylabel('Y lable', fontsize=15, labelpad=20)
plt.show()
```

```python

# 옵션
# legend 는 그래프에서 데이터가 나타내는 것을 표시하는 영역
# pyplot 객체.그래프명(..., label='범례명')
# pyplot 객체.legend(loc=(x,y) 또는 loc='위치')

import numpy as np
x = np.arange(1,5)
y = x**2

plt.plot(x,y,label='y=x^2')
plt.legend(loc='upper right')
plt.show()
```

```python
# 옵션
# xlim, ylim 표시 범위 지정 가능

x = np.arange(1,5) 
y = x**2

plt.plot(x,y)
plt.xlim([0,5])     # x축 표시 범위
plt.ylim([0,20])    # y축 표시 범위
plt.show()
```

```python
# 옵션
# pyplot 객체.axis([xmin, xmax, ymin, ymax])로 범위 지정

x = np.arange(1,5)
y = x**2

plt.plot(x,y)
plt.axis([0,5,0,20])
plt.show()
```

```python
# 여러 개 처리

x = np.arange(1,5)
plt.plot(x, x*2, marker='o', color='y', label='y=x*2')
plt.plot(x, x+3, marker='s', color='c', label='y=x+3')
plt.plot(x, x**2, marker='*', color='m', label='y=x**2')
plt.legend()
plt.show()
```
